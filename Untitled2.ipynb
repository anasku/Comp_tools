```python
import random                     
import matplotlib.pyplot as plt 
import math
from math import sqrt
from math import log
import random
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy
from scipy import stats

a = scipy.stats.cauchy.rvs(size=10,loc=30,scale=10)
b = scipy.stats.cauchy.rvs(size=100,loc=30,scale=10)
c = scipy.stats.cauchy.rvs(size=500,loc=30,scale=10)
d = scipy.stats.cauchy.rvs(size=1000,loc=30,scale=10)
alpha=np.arange(0.0, 60.0 , 0.1)


f=0
array=[]
for i in alpha:
    for s in a:
        f += np.log(10/(3.14*(10**2+(s-i)**2)))
    array.append(f)
    f=0    

plt.plot(alpha, array,"-",alpha[array.index(max(array))],max(array),"o")
plt.show()
print("alpha_ML = ",alpha[array.index(max(array))],"N=",len(c))

f=0
array=[]
for i in alpha:
    for s in b:
        f += np.log(10/(3.14*(10**2+(s-i)**2)))
    array.append(f)
    f=0    

plt.plot(alpha, array,"-",alpha[array.index(max(array))],max(array),"o")
plt.show()
print("alpha_ML = ",alpha[array.index(max(array))],"N=",len(b))

f=0
array=[]
for i in alpha:
    for s in c:
        f += np.log(10/(3.14*(10**2+(s-i)**2)))
    array.append(f)
    f=0    

plt.plot(alpha, array,"-",alpha[array.index(max(array))],max(array),"o")
plt.show()
print("alpha_ML =",alpha[array.index(max(array))], "N=",len(c))

f=0
array=[]
for i in alpha:
    for s in d:
        f += np.log(10/(3.14*(10**2+(s-i)**2)))
    array.append(f)
    f=0    

plt.plot(alpha, array,"-",alpha[array.index(max(array))],max(array),"o")
plt.show()
print("alpha_ML =",alpha[array.index(max(array))], "N=",len(d))



```
