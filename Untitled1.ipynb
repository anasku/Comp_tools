###### 1. Sampling $\pi$.

To estimate value of $\pi$ we are going to sample the uniformly distributed points in a square $-1<x<1$, $-1<y<1$ and count how many of them are inside of a unit circle. The probability for a given point to be inside of a circle is given by the formula 
\begin{equation}p=\frac{N_{inside}}{N_{total}}/.\end{equation}   
We define the random variable $S_i$ which takes the value $0$ if the point is outside the circle and $4$ if it is inside. Respectfully the probability of a r.v. to be $4$ is the same as a probability of a point hit inside a circle. 

1.The mean  and the variance of this r.v. can be calculated using the following formula

\begin{equation}
\begin{split}
m&=\sum_ip_iS_i\,,\\
\Delta&=\mathbb{E}(S_i^2)-m^2\,.
\end{split}
\end{equation}

2.For the first time we were throwing randomly 10000 points on a square. 
Using the formula for estimates of the mean and variance:
\begin{equation}
\begin{split}
\tilde m &= \sum_i S_i/N_{total}\,\\
\tilde \Delta &= \frac{1}{N_{total}-1}\sum_iS_i^2-\tilde m^2\,,
\end{split}
\end{equation}
we have found that estimates of mean and variance are equal to the mean and variance calculated using probability, so one can conclude that estimated are unbiased. 



```python
import random                     
import matplotlib.pyplot as plt   
from math import sqrt


num_trials = 10000                  
num_inside = 0
proba = 0
data_inside_x = []
data_inside_y = []
data_outside_x = []
data_outside_y = []
total = 0


for i in range(num_trials):
    x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)
    if x**2 + y**2 < 1.0: 
        num_inside += 1
        total += 4
        data_inside_x.append(x)
        data_inside_y.append(y)
        
    else:
        data_outside_x.append(x)
        data_outside_y.append(y)
        
proba = num_inside / num_trials
mean =  4.0 * proba
var = proba *16.0 - mean**2
estm = total / num_trials
estvar = (16 * num_inside / (num_trials - 1)) - estm*estm
var_est_mean = var / num_trials


print ("proba Si =4 :" , round(proba,3))
print ("mean =", round(mean,3))
print ("var=", round(var,3))
print ("est_mean=", round(estm,3))
print ("est_var=", round(estvar,3))
print ("var_est_mean =",round(var_est_mean,4))
print ("error = +-", round(sqrt(var_est_mean),4))

plt.plot(data_inside_x,data_inside_y,'.')
plt.plot(data_outside_x,data_outside_y,'x')
plt.axes().set_aspect('equal')
plt.show()
```

3.The variance of the estimate of the mean can be calculated using the formula:
\begin{equation}
var(\tilde m)=\frac{\Delta}{N_{total}}\,,
\end{equation}

then the mean value can be written using the estimate of the mean and the typical error made by this estimator:
\begin{equation}
m=\tilde m \pm \sqrt{\frac{\Delta}{N_{total}}}.
\end{equation}

4.Let us now repeat the procedure 1000 times for $N_{total}=10000\,,20000\,,25000\,$ and calculate the eastimates of mean and variance for each of them. One can see that the variance in case of $N_{total}=10000$ is the same as it has been predicted using the formula above. 

One can see from the histograms given below that the distribution becomes more peaked near the mean value 3.14 once we increase the number of trials. 





```python
import random
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#define a function counting the number 
#of points inside the circle for a given N
def direct_pi(N):
    num_inside = 0
    for i in range(N):
        x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)
        if x ** 2 + y ** 2 < 1.0:
            num_inside  += 1
    return num_inside

n_runs = 1000
n_trials = 10000
list_est = []

for run in range(n_runs):
    list_est.append(4*direct_pi(n_trials) / float(n_trials))
    
mean_est=sum(list_est)/len(list_est)
var =(np.std(list_est))**2
print("mean =", round(mean_est,5))
print("var=", round(var,4))

#varience = sum((mean_est - value)** 2 for value in list_est) / len(list_est)
#print("varience to compare=", varience)
   
plt.hist(list_est, bins = 50, normed=1, alpha=0.75)
plt.show() 
```
